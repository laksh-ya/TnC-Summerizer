## Defining and Testing Structured JSON Output Function


> Structured Output âœ…



```python
import os
from google import genai
from google.genai import types

def structuredTnC(test_prompt: str) -> None:
    """
    Generate a structured summary of Terms & Conditions using Gemini API.

    Args:
        test_prompt (str): The user-provided prompt or T&C text.
    """
    # Initialize the Gemini client
    client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))

    model = "gemini-2.0-flash"
    contents = [
        types.Content(
            role="user",
            parts=[ types.Part.from_text(text=test_prompt) ],
        ),
    ]

    # Define the desired output schema
    generate_content_config = types.GenerateContentConfig(
        response_mime_type="application/json",
        response_schema=genai.types.Schema(
            type=genai.types.Type.OBJECT,
            required=["terms_summary", "critical_watchpoints", "recommended_action"],
            properties={
                "terms_summary": genai.types.Schema(
                    type=genai.types.Type.ARRAY,
                    description="A crisp summary of the Terms & Conditions in exactly 10 bullet points, 1 line each.",
                    items=genai.types.Schema(type=genai.types.Type.STRING),
                ),
                "critical_watchpoints": genai.types.Schema(
                    type=genai.types.Type.OBJECT,
                    description="Exactly 2 critical risks, labeled Heading 1 and Heading 2, each followed by a short explanation.",
                    required=["Heading 1", "Heading 2"],
                    properties={
                        "Heading 1": genai.types.Schema(
                            type=genai.types.Type.STRING,
                            description="A short explanation of the first critical risk.",
                        ),
                        "Heading 2": genai.types.Schema(
                            type=genai.types.Type.STRING,
                            description="A short explanation of the second critical risk.",
                        ),
                    },
                ),
                "recommended_action": genai.types.Schema(
                    type=genai.types.Type.OBJECT,
                    description="Exactly 1 piece of advice under a heading.",
                    required=["Heading"],
                    properties={
                        "Heading": genai.types.Schema(
                            type=genai.types.Type.STRING,
                            description="One-line advice or caution for the user based on the risks.",
                        ),
                    },
                ),
            },
        ),
    )

    # Stream and print the response chunks
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        print(chunk.text, end="")
```

```python
structuredTnC(test_prompt)
```

## Defining & Testing the PDF Extraction Function


> Document Extraction âœ…



```python
import requests
import PyPDF2

def download_and_extract_pdf(url: str, test_pdf: str) -> str:
    """
    Downloads a PDF from the given URL to the local path `test_pdf`
    and returns its full text.

    :param url: The URL of the PDF to download.
    :param test_pdf: The local filename/path where the PDF will be saved.
    :return: The extracted text of the PDF.
    :raises RuntimeError: If the download fails.
    """
    # Download the PDF
    resp = requests.get(url)
    if resp.status_code != 200:
        raise RuntimeError(f"Failed to download PDF (status code {resp.status_code})")
    with open(test_pdf, "wb") as f:
        f.write(resp.content)

    # Extract text
    text = ""
    with open(test_pdf, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
```
